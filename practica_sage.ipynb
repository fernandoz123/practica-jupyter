```python
f(x) = (x^3+x^2+x)/(x^2-x-2)
p1(x) = x^3+x^2+x
p2(x) = x^2-x-2
p3(x) = p1.maxima_methods().divide(p2)[0]
p3

pf = plot(f(x), xmin = -8, xmax = 8, ymin = -20, ymax = 20, color = 'green', detect_poles = True, thickness = 1.5)
pp3 = plot(p3(x), xmin = -8, xmax = 8, ymin = -20, ymax = 20, color = 'red', linestyle = '--', legend_label = "Slant Asymptote")
pp3.set_legend_options(handlelength = 2)
v1 = line([(2, -30),(2,30)], color = 'black', linestyle = '--')
v2 = line([(-1, -30),(-1,30)], color = 'black', linestyle = '--')
r = pf+pp3+v1+v2

r.show()
```
